## Podstawowa analiza danych z rynku nieruchomości

## 1. Pobranie zestawu danych

```python
import pandas as pd
import os
from google.colab import files
import shutil

# Step 1: Upload the file to Google Colab
uploaded = files.upload()

# Step 2: Save the uploaded file to Google Drive
# Get the file name from the 'uploaded' dictionary
file_name = next(iter(uploaded))
destination_path = "/content/drive/MyDrive/Analizy/" + file_name

# Move the file to Google Drive (Replace 'path/to/save/' accordingly)
shutil.move(file_name, destination_path)

# Step 3: Determine file extension
file_extension = os.path.splitext(file_name)[1]

# Step 4: Load into Pandas DataFrame using appropriate reader
if file_extension == '.csv':
    df_loaded = pd.read_csv(destination_path)
elif file_extension == '.xlsx':
    df_loaded = pd.read_excel(destination_path)
elif file_extension == '.json':
    df_loaded = pd.read_json(destination_path)
# Add more conditions here for other file types
else:
    print(f'Unsupported file extension: {file_extension}')
```



## 2. Informacje dotyczące załadowanego pliku:


```python
df = df_loaded.copy(deep=True)
df.info()
```

Przykładowe 5 wpisów. Prezentowane jest tylko 20 pierwszych kolumn

```python
df.head(5)
```

## 3. Przetwarzanie danych wejściowych




### Wstępne czyszczenie danych

```python
# usuwanie zbędnych kolumn
drop_columns = ["Nr działki", "Arkusz mapy", "Funkcje budynku", "Struktura lokalu", "Pomieszczenia przynależne", "Kondygnacja", "Udział w nier. wspólnej", "Zdjęcia"]
keep_columns = []
len_drop_columns = len(drop_columns)
len_keep_columns = len(keep_columns)


if len_keep_columns > 0:
  df2 = df(keep_columns)

elif len_drop_columns > 0:
  df2 = df.drop(drop_columns, axis=1)

else:
  df2 = df.copy(deep=True)
```

```python
# zmiana nazw wybranych kolumn
column_name_mapping = {
    'old_name1': 'new_name1',
    'old_name2': 'new_name2',
    'old_name3': 'new_name3'
}
df2.rename(columns=column_name_mapping, inplace=True)
```

### Analiza korelacji

```python

numdf = df2.select_dtypes(include='number')
numdf.corr()
```
